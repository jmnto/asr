```python
def find_nclusters(hseg_file):
    idx_clusters = []
    nclusters = 0
    cluster_mask = ';;'
    for ids,i in enumerate(hseg_file):    
        if(i.find(cluster_mask) >= 0):
            nclusters = nclusters+1
            idx_clusters.append(ids)
    return nclusters,idx_clusters
```

```python
def seg_intervals(hseg_file):
    nclusters,idx_clusters = find_nclusters(hseg_file)
    cls_segs = []
    for i in range(nclusters-1):
        seg_list = hseg_file[(idx_clusters[i]+1):(idx_clusters[i+1])]
        cluster_segments = []
        for j in range(len(seg_list)):
            seg_j = seg_list[j]
            segj_fields = [m.start() for m in re.finditer(' ', seg_j)]
            cluster_segments.append([float(seg_j[(segj_fields[1]+1):segj_fields[2]]), float(seg_j[(segj_fields[1]+1):segj_fields[2]])+float(seg_j[(segj_fields[2]+1):segj_fields[3]])])
        cls_segs.append(cluster_segments)
    return cls_segs
    
```
